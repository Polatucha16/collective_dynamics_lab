```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
# import tensorflow as tf
import random
```

```python
# Podgląd trajektorii

start = 40
end = 50
partition = 1000
args = (0.6, 1.3, 1, 1)

def lotkavolterra_rhs(t, z, a, b, c, d):
    x, y = z
    return [a*x - b*x*y, c*x*y - d*y]

solution = solve_ivp(lotkavolterra_rhs,
                [start, end], # przedział czasowy
                [1, 1], # punkty startowe
                args=args, #parametry równania Lotki-Voltery
                dense_output=True,
                method='Radau')
t = np.linspace(start, end, partition)
dt = t[1]-t[0]
z = solution.sol(t)

plt.plot(*z)
plt.xlim(0,2.5)
plt.ylim(0,1.2)
plt.show()
print(f'Parameters (a,b,c,d)={args}, y0 = {[1, 1]}')
# grad_x, grad_y = np.gradient(z[0],t,edge_order=2), np.gradient(z[1],t,edge_order=2)
# z.shape
```

```python
# Parametry z równań liniowych z 3 kolejnyhc punktów trajektorii 
# z = [[x0,x1,x2,...,xN],[y0,y1,y2,...,yN]], t=[t0,t1,t2,...,tN]

x0, x1 = z[0,:-1], z[0,1:]
y0, y1 = z[1,:-1], z[1,1:]
x0y0 = z[0,:-1]*z[1,:-1]
x1y1 = z[0,1:]*z[1,1:]

iter_rhs = np.zeros(shape=(4,4,len(x0)), dtype=np.double)
iter_rhs[0,0], iter_rhs[0,1] = dt*x1, -dt*x1y1
iter_rhs[1,0], iter_rhs[1,1] = dt*x0, -dt*x0y0

iter_rhs[2,2], iter_rhs[2,3] = dt*x1y1, -dt*y1
iter_rhs[3,2], iter_rhs[3,3] = dt*x0y0, -dt*y0

iter_matrix = np.transpose(iter_rhs, axes=(2,0,1))
# iter_matrix
inverse_of_iter_rhs = np.linalg.inv(iter_matrix)
# np.linalg.det(inverse_of_iter_rhs)
# inverse_of_iter_rhs

numdiffs = np.transpose(
            np.array([[(z[0,2:]-z[0,1:-1])],
                      [(z[0,1:-1]-z[0,:-2])],
                      [(z[1,2:]-z[1,1:-1])],
                      [(z[1,1:-1]-z[1,:-2])]])
            ,axes=(2,0,1))
# numdiffs.shape
# inverse_of_iter_rhs[:-1].shape
# coef = np.squeeze((inverse_of_iter_rhs[:-1]@numdiffs).T)
coef = np.squeeze(np.linalg.solve(iter_matrix[:-1], numdiffs).T)
```

```python
# Historgram i statystyki

param_range = (0,np.amax(args)+1)

for i,color in enumerate(['b','y','g','r']):
    plt.hist(coef[i], bins='auto', alpha=0.5, facecolor=color, range=param_range)

plt.xlim(0, param_range[1] )
# plt.ylim(0, 0.03)

plt.show()

print("Mediany z znalezionych parametrów")
par_names = ['a','b','c','d']
for i,color in enumerate(['b','y','g','r']):
    print(f'median({par_names[i]}) = {np.median(coef[i])} ({par_names[i]} = {args[i]}), rel.diff = {np.abs((np.median(coef[i]-args[i]))/args[i])}')
```
